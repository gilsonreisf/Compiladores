```python
# (1) Seja a o primeiro símbolo de w$;
# (2) while { /*Repita indefinidamente*/
# (3) seja s o estado no topo da pilha;
# (4) if (ACTION [s,a] = shift t ) {
# (5) empilha t na pilha;
# (6) seja a o próximo símbolo da entrada;
# (7) }else if (ACTION [s,a] = reduce A-> β ) {
# (8) desempilha símbolos | β | da pilha;
# (9) faça o estado t agora ser o topo da pilha;
# (10) empilhe GOTO[t,A] na pilha;
# (11) imprima a produção A-> β ;
# (12) }else if (ACTION [s,a] = accept ) pare; /* a análise terminou*/
# (13) else chame uma rotina de recuperação do erro;

from goto import *
from action import *
from funcoes_auxiliares_sintatico import *
from regras import Regras

pilha = [0]

def PARSER(token):
  mainAction()
  regras = Regras()
  a = token # Token
  
  while True:
    s = pilha[-1]
    acao = action(s,a)
    t = int(acao[1:])
    print('Valor de t:', t)

    if('s' in acao):
      print('SHIFT')
      pilha.append(t)

      break # Retorna para o main pra chamar um novo token

    elif('R' in acao):
      print('REDUCE')

      A, B, regra = regras.retornaElementos(t)
      tamanhoDeBeta = len(B)

      print('Regra:', regra)
      print('Valor de A:', A)
      print('Valor de B:', B)


      for element in B:
        desempilhar(pilha)

      print('Pilha: ', pilha)
  
      t = pilha[-1]
      print('Valor de t após reduce:', t)

      pilha.append(int(goto(t,A)))
      print('Produção: ', regra)
      print('Pilha: ', pilha)

      
      break
      
    elif('a' in acao):
      print('ACCEPT')
      break

    else:
      print('ERRO')
      break
  


# action(4, 'inicio')
PARSER('inicio')
print('SAI DO WHILE')

```

```python
import pandas as pd
df_action = pd.read_csv('./tabelas/action.csv')


def isNaN(num):
    return num != num


def preencherErros(df, arquivoPath):
  colunas = df.columns
  qtdDeLinhas = len(df)
  for coluna in colunas:
    for index in range(qtdDeLinhas):
      if(isNaN(df[coluna][index])):
        df.loc[index, coluna] = 'e'+ str(index)

  df.to_csv(arquivoPath, index=False)


preencherErros(df_action, './tabelas/action.csv')
df_action
```
